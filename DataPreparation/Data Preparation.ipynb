## Correlation Methods

```python
import itertools

def linear_correlations(X, size=12):
    """
    Plots a graphical linear correlation matrix for every pair of features in the dataframe.
    
    X - A dataframe object
    size - Vertical and horizontal size of the plot
    """
    corr = X.corr()
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.grid()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    
    # adding heatmap scale
    cmap = matplotlib.cm.get_cmap('jet', 30)
    cax = ax.imshow(corr, interpolation="nearest", cmap=cmap)
    fig.colorbar(cax, ticks=np.arange(-1.0, 1.0, 0.1))
    
def polynomial_correlation_between_features(X, feature1, feature2):
    temp_data = X
    temp_data = temp_data.dropna(subset=[feature1, feature2], how='any')
    x = temp_data[feature1]
    y = temp_data[feature2]    
    poly_params = sp.polyfit(x, y, 2)
    yfit = poly_params[0]*(x**2)+poly_params[1]*(x**1)+poly_params[2]
    #print(yfit)
    yresid = y - yfit
    #print(yresid)
    SSresid = sum(pow(yresid,2))
    SStotal = len(y)* np.var(y)
    rsq = 1 - SSresid/SStotal    
    
    if rsq >= 0.7:
        print("Find polynomial correlations between: [%s] and [%s]" % (feature1, feature2))
        print("Polynom parameters = %s" % str(poly_params))
        print "RSQ = %s" % str(rsq)
        poly_2 = sp.poly1d(poly_params)
        xPoly = np.linspace(min(x), max(x), 100)  # Generate 100 x-coordinates from min(x) to max(x)
        yPoly = poly_2(xPoly)                     # Use the polynomial to calculate the y-coordinates
        P.plot(x, y,'o')
        P.plot(xPoly,yPoly,'r-')
        P.show()  

def polynomial_correlations(X, features):
    features_tuples = []
    lists = [sorted(f) for f in list(itertools.product(features, features))]
    for feature1, feature2 in lists:
        if feature1 != feature2:
            features_tuples.append((feature1, feature2))
    features_tuples = set(features_tuples)    
    for feature1, feature2 in features_tuples:
        polynomial_correlation_between_features(X, feature1, feature2)
```

## Categorical and Numerical Features Imputers

```python
from knnimpute import (knn_impute_few_observed, 
                       knn_impute_with_argpartition, 
                       knn_impute_optimistic,
                       knn_impute_reference)


class CategoricalImputer(object):
    def __init__(self, X, label, feature=None):
        """
        Impute missing values for categorical features by the label of the sample
        and selecting the most frequent value for that label group.
        
        This imputer doesn't handle features other of type object, and it can 
        handle a single feature imputation or multiple object typed features which 
        it retrives on its own.
        """
        self.label = label
        self.feature = feature
        self.X = X
        self.data_frame = X[[label, feature]] if self.feature else None
    
    def fit(self, x):
        return x.value_counts().index[0]
    
    def transform(self, x):
        return x.fillna(self.fit(x))
        
    def fit_transform(self):
        if self.feature:
            return self.data_frame.groupby(self.label)[self.feature].apply(lambda x: self.transform(x))
        else:
            columns_dictionary = dict(election_data.dtypes)
            categorical_features = [feature_name for feature_name, feature_type in columns_dictionary.iteritems() if feature_name != label_column_name and feature_type == "object"]
            for feature in categorical_features:
                self.data_frame = self.X[[self.label, feature]]
                self.X[feature] = self.data_frame.groupby(self.label)[feature].apply(lambda x: self.transform(x))
            return self.X
                
    
class NumericalKNNImputer(object):
    def __init__(self, X, strategy, k):
        """
        Impute missing values for numerical features by the strategy provided 
        as arguments.
        
        This imputer must receive numerical data
        """
        self.k = k
        self.data = X
        self.strategy = strategy
        if self.strategy == "optimistic":
            self.impute_function = knn_impute_optimistic
        elif self.strategy == "few_observed":
            self.impute_function = knn_impute_few_observed
        elif self.strategy == "argpartition":
            self.impute_function = knn_impute_with_argpartition
        elif self.strategy == "reference":
            self.impute_function = knn_impute_reference
        else:
            raise NotImplementedError
    
    def fit_transform(self):
        return self.impute_function(self.data, np.isnan(self.data), self.k)        
```

```python
def convert_categorical_features(X, categorical_features):
    """
    Convert the data object feature to numerical/boolean features for imputation
    
    X - the compleate data frame
    categorical_features - list of categorical features to convert their type
    """
    for feature in categorical_features:
        X[feature] = X[feature].astype("category")
        X[feature] = X[feature].cat.codes
    return X
    
    
election_data = convert_categorical_features(election_data, categorical_features + [label_column_name])
election_data.head()
```

## Outlier detection

```python
#Reference - http://colingorrie.github.io/outlier-detection.html

from sklearn import svm
from sklearn.covariance import EllipticEnvelope
from sklearn.ensemble import IsolationForest
#from sklearn.neighbors import LocalOutlierFactor

class OutlierDetector(object):
    def __init__(self, X, outliers_fraction=0.03):
        """
        Class initializer
        
        X - The dataset
        """
        self.X = X
        self.rng = np.random.RandomState(42)
        self.outliers_fraction = outliers_fraction
        self.n_samples = self.X[label_column_name].count()
        self.n_outliers = int(self.outliers_fraction * self.n_samples)
        self.n_inliers = self.n_samples - self.n_outliers
        self.n_neighbors = 50
        self.outlier_algorithms = { "One-Class SVM": svm.OneClassSVM(nu=self.outliers_fraction, kernel="rbf", gamma=0.1),
                                    "Robust Covariance": EllipticEnvelope(contamination=self.outliers_fraction),
                                    "Isolation Forest": IsolationForest(max_samples=self.n_samples, contamination=self.outliers_fraction, random_state=self.rng)}
    
    def detect_outliers_z_score(self, features, threshold=3):
        """
        Detect outliers samples of using z_score calculation.

        feature - continues feature name from the data set
        [threashold] - the threshold to mark samples as outlier 
        for their z_score value
        """
        samples_indexes = []
        for feature in features:
            data = self.X[feature]
            mean_x = np.mean(data)
            stdev_x = np.std(data)
            z_scores = [(x - mean_x) / stdev_x for x in data]
            samples_indexes.append(np.where(np.abs(z_scores) > threshold))
        return self.intersect_lists_values([outlier_tuple_list[0] for outlier_tuple_list in samples_indexes])

    def detect_outliers_modified_z_score(self, features, threshold=3.5):
        """
        Detect outliers samples of using a modified z_score calculation 
        using the median absolute deviation instead of the regular z_score 
        calculation.

        feature - continues features name from the data set
        [threashold] - the threshold to mark samples as outlier for their modified z_score value
        """
        samples_indexes = []
        for feature in features:
            data = self.X[feature]
            median_x = np.median(data)
            median_absolute_deviation_x = np.median([np.abs(x - median_x) for x in data])
            modified_z_scores = [0.6745 * (x - median_x) / median_absolute_deviation_x for x in data]
            samples_indexes.append(np.where(np.abs(modified_z_scores) > threshold))
        return self.intersect_lists_values([outlier_tuple_list[0] for outlier_tuple_list in samples_indexes])

    def detect_outliers_iqr(self, features, lower=15, upper=85):
        """
        Detect outliers samples of using interquartile range.
        Divides the data into four groups and the range between 
        the first and the third quartiles is the inside data and 
        each sample which is 1.5 times the below G1 or 1.5 times 
        above the G3 is an outside value.

        feature - continues feature name from the data set
        [threashold] - the threshold to mark samples as outlier for their modified z_score value
        """
        samples_indexes = []
        for feature in features:
            data = self.X[feature]
            quartile_1, quartile_3 = np.percentile(data, [lower, upper])
            iqr = quartile_3 - quartile_1
            lower_bound = quartile_1 - (iqr * 1.5)
            upper_bound = quartile_3 + (iqr * 1.5)
            samples_indexes.append(np.where((data > upper_bound) | (data < lower_bound)))
        return self.intersect_lists_values([outlier_tuple_list[0] for outlier_tuple_list in samples_indexes])

    def __get_detected_outlires_indexes(self, detected_outliers):
        """
        Convert the list of 1 and -1 values from the prediction to 
        the indexes of the samples for dropping from the dataset

        detected_outliers - List of 1 and -1 for each sample in the data
                            which mark the sample as inlier (1) and outlier (-1).
        """
        '''
        index_list = []
        for i in range(len(detected_outliers)):
            if detected_outliers[i] == -1:
                index_list.append(i)
        return np.array(index_list)
        '''
        outlier_indexes = list(np.where(np.array(detected_outliers) == -1)[0])
        return self.X.index[outlier_indexes]
        
    def __detect_outlires_by_algorithm(self, X, algorithm_name):
        """
        Generic function to call the algorithm for predicting 
        outlier samples
        
        X - The dataset
        algorithm_name - The name of the algorithm to operate
        """
        return self.__get_detected_outlires_indexes(self.outlier_algorithms[algorithm_name].fit(X).predict(X)) 
    
    def detect_outliers_one_class_svm(self, features):
        """
        One-Class SVM algorithm for detecting outlier samples 
        in the dataset     
        """
        return self.__detect_outlires_by_algorithm(self.X[features], "One-Class SVM")
        
    def detect_outliers_robust_covariance(self, features):
        """
        Robust Covariance algorithm using EllipticEnvelope for 
        detecting outlier samples in the dataset.
        """
        return self.__detect_outlires_by_algorithm(self.X[features], "Robust Covariance")
    
    def detect_outliers_isolationforest(self, features):
        """
        Isolation Forest algorithm for detecting outlier samples 
        in the dataset.             
        """
        return self.__detect_outlires_by_algorithm(self.X[features], "Isolation Forest")
        
    def intersect_lists_values(self, lists):
        sets = []
        main_set = set()
        if len(lists) > 0:
            for outlier_list in lists:
                if outlier_list.size > 1:
                    sets.append(set(outlier_list))
            if len(sets) > 0:
                main_set = sets[0]
                if len(lists) >= 2:
                    for other_set in sets[1:]:
                        main_set = main_set.intersection(other_set)
        return np.array(main_set)
    
    def clean_outlier_samples(self, outliers):
        """
        Drop given outliers samples indexes list from the dataset.
        
        outliers - Outlier samples indexes for dropping (numpay.array)      
        """
        if outliers.size > 1:
            return self.X.drop(labels=list(outliers))
        return self.X
        
```

# Normalization (scaling) 

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

class Normalizer(object):
    def __init__(self):
        return
        
    def min_max_normalization(self, X, uniform_features):
        """
        Normalize uniform distribution features in the data
        
        X - Data set
        uniform_features - List of uniform features
        """
        scaler = MinMaxScaler(feature_range=(-1, 1), copy=True)
        dataset = X.copy()
        for feature in uniform_features:
            dataset[feature] = scaler.fit_transform(dataset[feature].values.reshape(-1, 1))
        return dataset
    
    def z_score_normalization(self, X, gaussian_features):
        """
        Normalize gaussian distribution features in the data
        
        X - Data set
        gaussian_features - List of gaussian features
        """
        scaler = StandardScaler(copy=True, with_mean=True, with_std=True)
        dataset = X.copy()
        for feature in gaussian_features:
            dataset[feature] = scaler.fit_transform(dataset[feature].values.reshape(-1, 1))
        return dataset
```

```python
election_data_X = election_data.drop(label_column_name, axis=1).values
election_data_y = election_data[label_column_name].values
percentile = 55
print percentile

selector_classif = SelectPercentile(score_func=f_classif, percentile=percentile)
selector_mi = SelectPercentile(score_func=mutual_info_classif, percentile=percentile)

selector_classif.fit(election_data_X, election_data_y)
classif_scores = selector_classif.scores_
classif_scores /= classif_scores.max()
classif_features_percent = selector_classif.transform(election_data_X)
print classif_features_percent.shape

selector_mi.fit(election_data_X, election_data_y)
MI_scores = selector_mi.scores_
MI_scores /= MI_scores.max()
MI_features_percent = selector_mi.transform(election_data_X)
print MI_features_percent.shape

#Plot a see the significance of each feature evaluated using classif and MI score functions
plot_feature_selection_comparision(election_data_X, classif_scores, MI_scores)
```

```python
columns = np.asarray(election_data.drop(label_column_name, axis=1).columns)

support = np.asarray(selector_classif.get_support())
mi_seleceted_columns = columns[support]

support = np.asarray(selector_mi.get_support())
classif_seleceted_columns = columns[support]

print "Selected %d%% of features by SelectPercentile(f_classif)" % percentile
print len(classif_seleceted_columns)
print classif_seleceted_columns
print "Selected %d%% of features by SelectPercentile(mutual_info_classif)" % percentile
print len(mi_seleceted_columns)
print mi_seleceted_columns

filter_method_selected_features = list(set.intersection(set(classif_seleceted_columns), set(mi_seleceted_columns)))
print "Intersection between both methods of features selection"
print len(filter_method_selected_features)
print filter_method_selected_features

filtered_election_data_X = election_data[np.array(filter_method_selected_features)]
filtered_election_data_X.head()
```
